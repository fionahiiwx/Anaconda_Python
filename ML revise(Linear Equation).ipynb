```python
# define and compile the numeral network.
import tensorflow as tf
import numpy as np
from tensorflow import keras
# what's keras?
# the framework for defining neural network as a set of sequential layers.
```

```python
model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
xs = np.array([0.0, 2.0, 3.0], dtype = float)
ys = np.array([-3.0, 5.0, 9.0], dtype = float)
model.fit(xs, ys, epochs = 900) 
print(model.predict([4.0]))
```
